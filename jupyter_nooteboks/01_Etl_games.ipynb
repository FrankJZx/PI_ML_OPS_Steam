```python
# los generos se sacaron de la lista de generos antes de utilizar la columna tags para completar info. sacando: free to play, early 
# access
generos_juegos = ["Action", "Casual", "Indie", "Simulation", "Strategy", "RPG", "Sports", "Adventure", "Racing", "Massively Multiplayer", "Animation & Modeling", "Video Production", "Utilities", "Web Publishing", "Education", "Software Training", "Design & Illustration", "Audio Production", "Photo Editing", "Accounting"]

# Elimina los elementos duplicados de la lista
generos_juegos = list(set(generos_juegos))

print(generos_juegos)
```

```python
# Convertir la lista en un conjunto
lista_comparacion = set(generos_juegos)

# Filtrar el DataFrame por las filas que coinciden con el conjunto
df_games = df_exploded.loc[df_exploded['genres'].isin(lista_comparacion)]

df_games
```

ahora pasaria a la normalizacion de las demas columnas para que no tengan nulos/vacios/etc, para el correcto funcionamiento del modelo
